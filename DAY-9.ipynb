```python
num = [1, 3, 5, 7, 9, 4,]

largest = num[0]

for num in num :
    if num > largest :
        largest = num

print('largest number :',largest)
```

```python
num = [2, 34, 85, 73, 1]

total = 0
for num in num:
    total += num

print('sum:' , total)
```

```python
nums = [1, 2, 2, 3, 3, 4,]
uni = []

for num in nums:
    if num not in uni:
        uni.append(num)

print(uni)
```

```python
word = input('inter a word :')
ch = input('inter a character :')

count = 0

for i in word:
    if i == ch:
        count += 1

print(count)
```
