```python
n = int(input("Enter a number: "))

count = 0

for i in range(1, n+1):
    if n % i == 0:
        count += 1

if count == 2:
    print("Prime")
else:
    print("Not Prime")
```

```python
n = int(input("Enter a number: "))

for num in range(2, n+1):
    count = 0

    for i in range(1, num+1):
        if num % i == 0:
            count += 1

    if count == 2:
        print(num, end=" ")
```
