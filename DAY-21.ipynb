```python
num = input('Enter a words')

if num == num[::-1]:
        print(num,'is a palindrome')
else:
        print(num,'is not a palindrome')
```

```python
x = [1, 2, 3, 4]
y = [3, 4, 5, 6]

for i in x:
    if i in y:
        print(i, 'common')
```
