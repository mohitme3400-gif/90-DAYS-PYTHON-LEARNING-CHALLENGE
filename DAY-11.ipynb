```python
n = int(input("Enter number of rows: "))

for i in range(1, n+1):
    for j in range(i):
        print("*", end="")
    print()
```

```python
sentence = input("Enter a sentence: ")

words = sentence.split()
words.reverse()

result = " ".join(words)

print(result)
```

```python
text = input("Enter a string: ")

letters = 0
digits = 0
spaces = 0

for ch in text:
    if ch.isalpha():
        letters += 1
    elif ch.isdigit():
        digits += 1
    elif ch.isspace():
        spaces += 1

print("Letters =", letters)
print("Digits =", digits)
print("Spaces =", spaces)
```
