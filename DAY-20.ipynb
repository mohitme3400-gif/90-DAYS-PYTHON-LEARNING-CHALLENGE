```python
num = [23, 45, 65, 72, 24]
smallest = num[0]

for i in num:
    if i < smallest:
        smallest = i

print(smallest, 'is a smallest number')
```

```python
num = int(input("Enter number: "))

original = num
sum = 0

while num > 0:
    digit = num % 10
    sum += digit ** 3
    num = num // 10

if original == sum:
    print("Armstrong Number")
else:
    print("Not Armstrong")
```

```python
num = int(input('Enter Number'))
count = 0

while num > 0:
    num = num // 10
    count += 1

print('Digits: ',count)
```
