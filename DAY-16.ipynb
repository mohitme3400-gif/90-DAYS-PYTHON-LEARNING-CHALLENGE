```python
def is_prime(n):
    if n < 2:
        return "Not Prime"

    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return "Not Prime"

        return "Prime"
```

```python
is_prime(19)
```

```python
def find_gcd(a, b):
    gcd = 1
    for i in range(1, min(a, b)+1):
        if a % i == 0 and b % i == 0:
            gcd = i
    return gcd
```

```python
find_gcd(12, 13)
```
