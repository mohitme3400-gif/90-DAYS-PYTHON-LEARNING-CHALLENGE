```python
import pandas as pd
```

```python
name = ['mohit', 'rahul', 'Aman']
marks = [80, 45, 60]

for i in range(len(name)):
    print('Name', name[i])
    print('Marks', marks[i])

    if marks[i] >= 50:
        print('result: pass')
    else:
        print('result: fail')
```
