```python
num = input("Enter a number: ")
sum = 0

for i in num:
    sum += int(i)

print("Sum of digits:", sum)
```

```python
n = int(input("Enter a number: "))

for i in range(10, 0, -1):
    print(n, "x", i, "=", n * i)
```

```python
num = input("Enter a number: ")

even = 0
odd = 0

for i in num:
    if int(i) % 2 == 0:
        even += 1
    else:
        odd += 1

print("Even digits:", even)
print("Odd digits:", odd)
```
