```python
num = [12, 45, 56, 72, 64]
largest = num[0]

for i in num:
    if i > largest:
        largest = i

print(largest, 'is a largest')
```

```python
year = int(input("Enter year: "))

if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print("Leap Year")
else:
    print("Not Leap Year")
```
