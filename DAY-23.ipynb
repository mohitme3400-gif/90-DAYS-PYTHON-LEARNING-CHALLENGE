```python
num = [1, 2, 3, 4, 5]
x = []

for i in num:
    x = [i] + x

print(x)
```

```python
num = [10, 5, 20, 25]
maxnumber = num[0]
minnumber = num[0]

for i in num:
    if i > maxnumber:
        maxnumber = i
    if i < minnumber:
        minnumber = i

print(minnumber, 'is a smallest number')
print(maxnumber, 'is a largest number')
```

```python
num = [13, 34, 45, 67, 13, 12]
even = []
odd = []

for i in num:
    if i % 2 == 0:
        even.append(i)
    else:
        odd.append(i)

print(even, 'even number')
print(odd, 'odd number')
```

```python
def sum_digits(num):
    total = 0
    for ch in str(num):
        total += int(ch)
    return total
print(sum_digits(123))
```
